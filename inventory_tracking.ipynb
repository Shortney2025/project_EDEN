```python
def load_items():
    """
    Prompts the user to enter items, quantities, and prices,
    and returns a list of dictionaries, each representing an item.

    Input format: 'item_name,quantity,price' (e.g., 'milk,2,6.89').
    Enter 'done' to finish input.
    Handles invalid input formats, non-numeric quantities/prices,
    and negative quantities/prices gracefully by skipping the line.
    """
    items = []
    while True:
        user_input = input("Enter item,qty,price (or 'done' to finish): ")
        if user_input.lower() == 'done':
            break

        parts = user_input.split(',')
        if len(parts) != 3:
            print("Invalid input format. Please use 'item,qty,price'.")
            continue

        item_name = parts[0].strip()

        # Parse quantity
        qty_str = parts[1].strip()
        try:
            qty = int(qty_str)
        except ValueError:
            print(f"Invalid quantity '{qty_str}' for '{item_name}', skipping line.")
            continue

        # Parse price
        price_str = parts[2].strip()
        try:
            price = float(price_str)
        except ValueError:
            print(f"Invalid price '{price_str}' for '{item_name}', skipping line.")
            continue

        if qty < 0 or price < 0:
            print(f"Quantity ({qty}) and price (${price:.2f}) cannot be negative for '{item_name}', skipping line.")
            continue

        items.append({"item": item_name, "qty": qty, "price": price})
    return items

# Example of how to use the function:
inventory = load_items()
print(inventory)
```
